```python
print(df['Plant_Health_Status'].unique())
df['Plant_Health_Status']=df['Plant_Health_Status'].map({'Healthy':0,'Moderate Stress':1, 'High Stress':2})
print(df['Plant_Health_Status'])
```

```python
#splitting the data for training and testing
#input data
x=df.drop('Plant_Health_Status',axis=1)  #labelling the data
y=df['Plant_Health_Status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3) #20%=0.2
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ss= StandardScaler()

x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
x_train
```

```python
#model implementation

from sklearn.linear_model import LogisticRegression  
# LogisticRegression is a class, to use a class first we have to create object

lr= LogisticRegression()
#whatever the objct name we give it is model(lr)

#training model
lr.fit(x_train,y_train)
```

```python
#testing model(predicting results)
#predicts 30% of x_test for testing

y_pred= lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

#ANY VALUES WHICH IS IN 2D WE CAN USE HEATMAP

cm= confusion_matrix(y_test,y_pred)
print("Confusion Matrix",cm)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, #values in heatmap
            fmt="d", #numerical values in heatmap(ex:104)
            cmap="Blues", cbar=False, xticklabels=['Predicted class 0', 'Predicted class 1','Predicted class 2'],  #machine predicted value
            yticklabels=['Actual class 0', 'Actual class 1','Actual class 2'])  #target value model needs to predicted
#add labels
plt.xlabel('Predicted label')
plt.ylabel('true label')
plt.title('confusion matrix')
plt.show()
            
accuracy= accuracy_score(y_test,y_pred)
print("Accuracy Score",accuracy)
```

```python
#model implementation

from sklearn.tree import DecisionTreeClassifier 

dtc= DecisionTreeClassifier()
#whatever the objct name we give it is model(dtc)

#training model
dtc.fit(x_train,y_train)
```

```python
#testing model(predicting results)
#predicts 30% of x_test for testing

y_pred= dtc.predict(x_test)
y_pred
```

```python
#model implementation

from sklearn.ensemble import RandomForestClassifier 

rfc= RandomForestClassifier()
#whatever the objct name we give it is model(rfc)

#training model
rfc.fit(x_train,y_train)
```

```python
def predict_health(Soil_Moisture=15.336156, Ambient_Temperature=21, Soil_Temperature=15,Humidity=55, Light_Intensity=650, Soil_pH=29, Nitrogen_Level=50, Phosphorus_Level=25,
                 Potassium_Level=30, Chlorophyll_Content=45, Electrochemical_Signal=1):
    temp_array = list()

    #Soil_Moisture
    temp_array=temp_array+[Soil_Moisture]

    #Ambient_Temperature
    temp_array=temp_array+[Ambient_Temperature]
    temp_array=temp_array+[Soil_Temperature]

    #Humidity
    temp_array=temp_array+[Humidity]

    #Light_Intensity
    temp_array=temp_array+[Light_Intensity]

    #Soil_pH
    temp_array=temp_array+[Soil_pH]

    #Nitrogen_Level
    temp_array=temp_array+[Nitrogen_Level]

    #Phosphorus_Level
    temp_array=temp_array+[Phosphorus_Level]

    #Potassium_Level
    temp_array=temp_array+[Potassium_Level]

    #Chlorophyll_Content
    temp_array=temp_array+[Chlorophyll_Content]

    #Electrochemical_Signal
    temp_array=temp_array+[Electrochemical_Signal]

#converting into numpy array
    temp_array= np.array([temp_array])
    print(temp_array)
    temp_array=ss.transform(temp_array)
    pred=rfc.predict(temp_array)
    print(pred)
    if pred[0]==0:
        result='Healthy Plant'
    elif pred[0]==1:
        result='Moderate Stress'
    else:
        result="High Stress"

    #prediction
    return result
        

```
